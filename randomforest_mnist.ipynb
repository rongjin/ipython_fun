```python
%matplotlib inline
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
traindata = pd.read_csv("../input/train.csv")
testdata = pd.read_csv("../input/test.csv")
m,n = traindata.shape
train_X = traindata.ix[:,1:].as_matrix()
train_Y = traindata['label'].as_matrix()
train_X.shape
```

```python
clf = RandomForestClassifier(n_estimators=100,random_state=0)
clf.fit(train_X,train_Y)
test_X = testdata.as_matrix()

```

```python
pred = clf.predict(test_X)
pred = np.c_[np.arange(1,test_X.shape[0]+1), pred]  
result = pd.DataFrame(np.matrix(pred))
result.to_csv("rf.csv",header = ["ImageId", "Label"],index=False)
```
